```python
from transformers import pipeline
```

```python
emotionModel = pipeline(
    "text-classification",
    model="j-hartmann/emotion-english-distilroberta-base",
    truncation=True,  # Truncate long sequences
    max_length=512     # Set maximum sequence length
)
```
